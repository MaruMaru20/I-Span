# [dbo].[vw_vacancy_rate]

```sql
/****** Object:  View [dbo].[vw_vacancy_rate]    Script Date: 2024/12/3 下午 12:47:51 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create view [dbo].[vw_vacancy_rate] as
select concat (round((
	select count(*) as '空屋'
	from House left join live
		on House.hid = live.hid
	where uid is null
)
/cast((select count(*)  from house) as float), 3)* 100, '%') as 'rate'
GO
```

# [dbo].[login]

```sql
/****** Object:  StoredProcedure [dbo].[login]    Script Date: 2024/12/3 下午 12:47:51 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[login]
	@uid nvarchar(50),
	@pwd nvarchar(50)
as
begin
	declare @isok tinyint = 0
	select @isok = count(*) from userinfo where uid = @uid and password =@pwd
	select @isok as 'status'
end
GO
```
